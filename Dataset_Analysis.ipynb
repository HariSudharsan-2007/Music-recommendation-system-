```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load the dataset
df = pd.read_csv(r'')
X = df.drop('label', axis=1)
y = df['label']

# Set plot style
sns.set_theme(style="whitegrid")

# Create label distribution plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='label', order=['POSITIVE', 'NEUTRAL', 'NEGATIVE'], palette={'POSITIVE': '#2ecc71', 'NEUTRAL': '#95a5a6', 'NEGATIVE': '#e74c3c'})
plt.title('Distribution of Emotional States')
plt.xlabel('Emotion')
plt.ylabel('Number of Samples')
plt.savefig('label_distribution.png', bbox_inches='tight')
plt.close()

# Find top 3 features with highest variance
variances = X.var().sort_values(ascending=False)
top_features = variances.head(3).index.tolist()

# Create feature impact boxplots
plt.figure(figsize=(18, 6))
for i, feature in enumerate(top_features, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x='label', y=feature, data=df, order=['POSITIVE', 'NEUTRAL', 'NEGATIVE'], palette={'POSITIVE': '#2ecc71', 'NEUTRAL': '#95a5a6', 'NEGATIVE': '#e74c3c'})
    plt.title(f'Impact of Feature: {feature}')
    plt.xlabel('Emotion')
    plt.ylabel('Feature Value')
plt.tight_layout()
plt.savefig('feature_impact.png', bbox_inches='tight')
plt.close()

# Scale features for dimensionality reduction
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensions to 50
pca = PCA(n_components=50, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Apply tSNE for 2D visualization
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_pca)

# Create a dataframe for tSNE plotting
tsne_df = pd.DataFrame({'Dimension 1': X_tsne[:, 0], 'Dimension 2': X_tsne[:, 1], 'Label': y})

# Plot tSNE
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Dimension 1', y='Dimension 2', hue='Label', 
                data=tsne_df, palette={'POSITIVE': '#2ecc71', 'NEUTRAL': '#95a5a6', 'NEGATIVE': '#e74c3c'}, 
                alpha=0.8, edgecolor=None)
plt.title('tSNE Visualization of Class Separability')
plt.legend(title='Emotion')
plt.savefig('tsne_plot.png', bbox_inches='tight')
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
df = pd.read_csv(r'')
X = df.drop('label', axis=1)
y = df['label']

# Train the Random Forest Model
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X, y)

# Extract Feature Importances
importances = rf.feature_importances_

# Map scores to their corresponding feature names
feature_importance_df = pd.DataFrame({
    'Feature_Name': X.columns,
    'Importance_Score': importances
})

# Sort the values from highest to lowest score
feature_importance_df = feature_importance_df.sort_values(by='Importance_Score', ascending=False)

# Get the top 15 features
top_15_features = feature_importance_df.head(15)

# Display the top 15 most important features in the console
print("Top 15 Most Important Brainwave Features:")
print(top_15_features.to_string(index=False))

# Set the style for the plot
sns.set_theme(style="whitegrid")

# Create a horizontal bar chart for the top 15 features
plt.figure(figsize=(10, 8))
sns.barplot(
    x='Importance_Score', 
    y='Feature_Name', 
    data=top_15_features, 
    palette='viridis'
)

# Add titles and labels
plt.title('Top 15 Most Important Brainwave Features', fontsize=16, fontweight='bold')
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Feature Name', fontsize=12)

# Save and show the plot
plt.tight_layout()
plt.savefig('top_15_feature_importance.png', bbox_inches='tight')
plt.show()
```
